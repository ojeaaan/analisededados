```python
eventos = {'m1': list(m1),
           'm2': list(m2),
           'm3': list(m3),
           'm4': list(m4),
           'm5': list(m5)}
moedas = pd.DataFrame(eventos)
df = pd.DataFrame(data=['Cara', 'Coroa'],
                  index=['c', 'C'],
                  columns=['Faces'])
for item in moedas:
    df = pd.concat([df, moedas[item].value_counts()],
                   axis=1)
df
```
