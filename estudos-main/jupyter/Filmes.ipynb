```python
tmdb["original_language"].value_counts().to_frame()
```

```python
#separando os index
contagem_de_lingua = tmdb["original_language"].value_counts().to_frame().reset_index()
contagem_de_lingua.columns = ["original_language","total"]
contagem_de_lingua.head()
```

```python
import seaborn as sns
```

```python
#sns.barplot(data = contagem_de_lingua)
contagem_de_lingua = sns.barplot(x="original_language", y="total", data = contagem_de_lingua)
contagem_de_lingua.figure.set_size_inches(25, 6)
```

```python
contagem_de_lingua = sns.catplot(x="original_language", kind="count", data=tmdb)
contagem_de_lingua.figure.set_size_inches(25, 6)

#aqui é uma forma mais simplificada de fazer o que foi feito em cima, de separar por index, pegar os valores
#pegar o titulo, aqui já faz direto, PORÉM é necessário ter a versão 0.9 para cima!
```

*Aqui irá mostrar só o do inglês*

```python
total_por_lingua = tmdb['original_language'].value_counts()
total_por_lingua.loc['en']
```

```python
total_por_lingua = tmdb['original_language'].value_counts()
total_geral = total_por_lingua.sum()
total_de_ingles = total_por_lingua.loc['en']
total_do_resto = total_geral - total_de_ingles
print(total_de_ingles,total_do_resto)
```

```python
dados = {
    'lingua': ['ingles','outros'],
    'total':[total_de_ingles,total_do_resto]
}
dados
```

```python
pd.DataFrame(dados)
```

```python
sns.barplot(x="lingua",y="total",data = dados)
```

**ALTERAR O TAMANHO COM O SNS ORIGINAL**

```python
total_lingua_otros_filmes = tmdb.query("original_language != 'en'")
sns.catplot(x="original_language",kind="count",data = total_lingua_otros_filmes, aspect=4)
```

No order é possível colocar do maior para o menor, como vemos que nos códigos de cima já veio hierarquizado os valores então colocamos no order

```python
total_lingua_otros_filmes = tmdb.query("original_language != 'en'")
sns.catplot(x = "original_language", kind="count", data = total_lingua_otros_filmes,palette="GnBu_r", aspect=4, order = total_lingua_otros_filmes_ord.index)
#GnBu_r ou GnBu_d - o que acontece o R para reverse
```
